### Import libraries

```python
import numpy as np, cv2
from skimage.io import imsave
from skimage.measure import block_reduce
from torch.nn import MaxPool2d
from torch import from_numpy
from matplotlib import pyplot as plt
from matplotlib.image import imread
```

### User-defined Functions (UDFs)

```python
def plot(
    imgs : list,
    titles : list = ["ORIGINAL", "EDITED"],
    figsize : tuple = (8, 5),
    cmaps : list = [None, None],
    *args, **kwargs
) -> None:

    if len(imgs) > 3:
        raise ValueError("Max. total image = 3.")

    if len(imgs) != len(titles):
        raise ValueError("'imgs' should be in the same length with 'titles'.")

    my_figure = plt.figure(figsize=figsize)
    # NOTE: figsize=(width/horizontally, height/vertically)

    fig_ids = [121, 122] if len(imgs) == 2 else [131, 132, 133]

    for idx, img, title, c in zip(fig_ids, imgs, titles, cmaps):
        m = my_figure.add_subplot(idx)
        plt.imshow(img, cmap=c)
        plt.axis("off")
        plt.title(title)

    plt.tight_layout()
```

```python
def apply_clahe_rgb(img, clipLimit=2, tileGridSize=(10, 10)):
    # LAB color model where L = Lightness, a = green-red, b = blue-yellow
    light, a, b = cv2.split(cv2.cvtColor(img, cv2.COLOR_BGR2LAB)) # convert to LAB color model
    light_clahe = cv2.createCLAHE(clipLimit=clipLimit, tileGridSize=tileGridSize).apply(light) # apply CLAHE for lightness (L) component only
    return cv2.cvtColor(cv2.merge((light_clahe, a, b)), cv2.COLOR_LAB2BGR) # return with reconvert to BGR color model
```

### Max Pooling (using Scikit-image)

```python
BLOCK_SIZE = (4, 4, 1)
```

```python
img_edited = block_reduce(img, block_size=BLOCK_SIZE, func=np.max)
plot(imgs=[img, img_edited])
```

### Max Pooling (using PyTorch)

```python
# convert numpy-based image to torch.Tensor type
img_torch = from_numpy(img).permute(2, 0, 1).float()

# prepare Max Pooling operation from PyTorch
max_pooling = MaxPool2d(kernel_size=9, stride=2)

# perform Max Pooling operation
img_torch_edited = max_pooling(img_torch).permute(1, 2, 0).int().numpy()
plot(imgs=[img, img_torch_edited])
```

### Min Pooling and Average Pooling

```python
img_edited_min = block_reduce(img, block_size=BLOCK_SIZE, func=np.min) # TODO: Recall the block_reduce(), but with func=np.min
plot(imgs=[img, img_edited_min])
```

```python
img_edited_avg = block_reduce(img, block_size=BLOCK_SIZE, func=np.mean).astype(int) # TODO: Recall the block_reduce().astype(int), but with func=np.mean
plot(imgs=[img, img_edited_avg])
```

[ ANSWER HERE ]

**1. Min Pooling**
- Operasi ini memilih nilai minimum dari setiap jendela (kernel) dlm gambar saat melakukan subsampling atau pengurangan dimensi.
- Hasilnya adalah gambar yang menyoroti nilai minimum di setiap jendela dan cenderung mempertahankan fitur yang paling gelap atau rendah dalam gambar.
- Ini akan menghasilkan gambar dengan kontras yang tinggi di area-area yang memiliki perbedaan intensitas yang signifikan.

**2. Average Pooling**
- Operasi ini mengambil nilai rata-rata dari setiap jendela dalam gambar saat melakukan subsampling atau pengurangan dimensi.
- Hasilnya adalah gambar yang lebih halus dan mewakili nilai rata-rata intensitas dalam setiap jendela.
- Ini akan menghasilkan gambar yang lebih halus dan kurang memiliki perubahan kontras yang tajam.


***KESIMPULAN***

Jadi, perbedaan utamanya yaitu Min Pooling mempertahankan nilai minimum dalam setiap jendela, sedangkan Average Pooling mengambil rata-rata nilai intensitas dalam jendela. Dalam kasus gambar dengan perbedaan intensitas yang signifikan, Min Pooling akan menghasilkan perubahan yang lebih dramatis dibandingkan dengan Average Pooling.

### CLAHE
Now imagine you have a better approach to compete Apple's Photonic Engine and Samsung's Adaptive Tetra-squared Pixel technologies using **Contrast Limited Adaptive Histogram Equalization (CLAHE)**. How's the result?

```python
img_clahe = apply_clahe_rgb(img) # TODO: Call the function apply_clahe_rgb() and pass the variable image of "photo1.jpeg"
plot(imgs=[img, img_edited, img_clahe], titles=["ORIGINAL", "MAX-POOLING", "CLAHE"], figsize=(12, 5), cmaps=[None]*3)
```

[ ANSWER HERE ]

**1. Pengaturan Kontras yang Lebih Halus**
- CLAHE menghasilkan perubahan kontras yang lebih halus daripada Max Pooling. Ini membuatnya lebih cocok untuk gambar yang memerlukan peningkatan kontras yang lebih alami.


**2. Pemahaman Intensitas Gambar**
- CLAHE berfokus pada pemahaman intensitas piksel dalam gambar. Ini berarti CLAHE akan berusaha untuk mempertahankan kontras antara berbagai bagian gambar, menjadikan detail lebih mudah dilihat. Di sisi lain, Max Pooling mengambil nilai maksimum dalam jendela, yang dapat mengakibatkan hilangnya sebagian besar informasi detail.

**3. Kontrol Terbatas Kontras**
- CLAHE memiliki parameter clipLimit yang memungkinkan kita dlm mengontrol tingkat kontras yang diterapkan pada gambar. Ini memungkinkan kita untuk membatasi kontras yang ditingkatkan agar tidak terlalu ekstrem. Di sisi lain, Max Pooling hanya mengambil nilai maksimum dalam jendela, yang dapat menghasilkan perubahan kontras yang sangat drastis.


***KESIMPULAN***

Jadi, keuntungan utama menggunakan CLAHE adalah kemampuannya untuk meningkatkan kualitas gambar dalam kondisi minim cahaya dengan mempertahankan kontras dan detail yang penting, sambil tetap memiliki kontrol atas tingkat kontras yang diterapkan pada gambar.
